## Manual Binning

Manual binning dilakukan ketika kita tahu aturan dan rentang yang ingin dibuat

```python
bin_goal_interval = [0, 1, 2, 3, 4]
category = ['Sangat Sedikit', 'Sedikit', 'Sedang', 'Banyak']
```

```python
df['bin_goal_interval'] = pd.cut(df['xG'], bins=bin_goal_interval, labels=category)
```

Binning pada xG, xG adalah pengukuran kualitas sebuah peluang terciptanya sebuah goal

```python
print(df[['teamId','xG', 'bin_goal_interval']].head(20)) # 20 tim yang bergabung di EPL pada tahun 2020
```

## Binning dengan Linspace

Binning dengan mencari tahu ukuran minimum dan maksimum dari sebaran datanya

```python
bin_linspace = np.linspace(min(df['xG']), max(df['xG']), 5)
```

```python
print(bin_linspace)
```

```python
df['binned_2'] = pd.cut(df['xG'], bins=bin_linspace, labels=category, include_lowest=True)
```

```python
print(df[['teamId','xG', 'bin_goal_interval', 'binned_2']].head(20)) # 20 tim yang bergabung di EPL pada tahun 2020
```

## Binning dengan Quantile

Digunakan ketika kita tidak tahu aturannya, rumusnya, atau tidak paham tentang rentang datanya

```python
df['binned_3'] = pd.qcut(df['xG'], 4)
```

```python
print(df[['teamId','xG', 'bin_goal_interval', 'binned_2', 'binned_3']].head(20)) # 20 tim yang bergabung di EPL pada tahun 2020
```

```python
df['binned_3'] = pd.qcut(df['xG'], q=4, labels=category)
print(df[['teamId','xG', 'binned_3']].head(20))
```

# Encoding
Kualitatif menjadi Kuantitatif dengan menggunakan encoding, pengubahan data kategorik menjadi nilai numerik yang unik, contohnya 1, 2, 3 ataupun dengan biner 1 dan 0

## One-Hot Encoding
Teknik untuk mengubah data kategorik menjadi data numerik dengan nilai biner

```python
from sklearn.preprocessing import OneHotEncoder
```

contohnya, ingin mengetahui hari apa saja tim-tim yang ada di epl bertanding

```python
y = OneHotEncoder(sparse_output=False)
y = y.fit_transform(df[['matchDay']])
```

```python
data_y = pd.DataFrame(y)
data_y
```

Match Day for Each Team

```python
header = df['teamId'].sort_values()
data_y.index = header + '_MatchDay'
data_y.columns = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
data_y
```
